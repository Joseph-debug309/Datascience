```python
import pandas

data = pandas.read_csv("datasets/supermarket_data.csv")
data
```

```python
# Use the describe function
data.describe()
```

```python
# we've noticed that the column for item price contains 3 empty slots
# We need to fill these empty slots with values. in this case we shall fill with the mean of the column
data["itemprice"] = data["itemprice"].fillna(55)

# Describe the data again
data.describe()
```

```python
# Get data from a given column
data.groupby("gender").size()
```

```python
# combining multiple columns
data.groupby(["city", "payment"]).size()
```

```python
# Import the modules
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Below is a countplot showing the distribution of males to female
sns.countplot(x="gender", data=data, palette="winter")
plt.title("Gender distribution against total count")
plt.ylabel("Total count of individuals")
```

```python
# Below is a countplot showing the distribution of males to female
# Which product category appears the most?
sns.countplot(x="product", data=data, palette="winter")
plt.xticks(rotation = 90)
plt.title("Distribution of products")
```

```python
# below is a count plot that has two columns
sns.countplot(x="gender", hue="city", palette="summer", data=data)
```

```python
# Below is a pie plot
data["payment"].value_counts().plot.pie(autopct ='%1.1f%%')
```
